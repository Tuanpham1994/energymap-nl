```python
#Import multiple data file
import glob
import os
from pathlib import Path

CURRENT_PATH = Path().resolve()
DATA_PATH = CURRENT_PATH / 'data'
ELECTRICITY_PATH = DATA_PATH / 'Electricity'
```

```python
all_files_energy = ELECTRICITY_PATH.glob("*.csv")
```

```python
all_files_energy
```

```python
#li_energy = [pd.read_csv(filename, index_col = None, header = 0) for filename in all_files_energy]
li_energy = []
for csv in all_files_energy:
    frame = pd.read_csv(csv)
    frame['filename'] = os.path.basename(csv)
    li_energy.append(frame)
li_energy[0][0:5]
```

```python
df_energy = pd.concat(li_energy, axis = 0, ignore_index = True)
df_energy
```
